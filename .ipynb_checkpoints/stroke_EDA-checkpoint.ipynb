```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('stroke.csv')
df.head()
len(df)
```

```python
# exclude children
df = df[df['age'] > 16]
df.head()
```

```python
df['smoking_status'].unique()
```

```python
# replace 
# 0: never smoked
# 1: formerly smoked
# 2: smokes
# 3: unknown
#all_xray_df['pneumonia_class'] = all_xray_df.apply(lambda x: "P" if x.Pneumonia == 1.0 else "N", axis=1)

df['smoking_status'] = df.apply(lambda x: 0 if x['smoking_status'] == 'never smoked' else 1 if x['smoking_status'] == 'formerly smoked'
                      else 2 if x['smoking_status'] == 'smokes' else 3, axis = 1)
```

```python
# Male: 0, Female: 1
df['gender'] = df.apply(lambda x: 0 if x['gender'] == 'Male' else 1, axis = 1)
df.head()
```

```python
# replace
# 0: never worked, 1: goverment, 2: self-employed, 3: private
df['work_type'].unique()
#df['work_type'].value_counts().plot(kind='bar')
```

```python
df['work_type'] = df.apply(lambda x: 0 if x['work_type'] == 'Never worked' else 1 if x['work_type'] == 'Govt_job' 
                           else 2 if x['work_type'] == 'Self-employed' else 3, axis=1)
df.head()
```

```python
df['Residence_type'].unique()
df['Residence_type'].value_counts().plot(kind='bar')
```

```python
# 0: urban, 1: rural
df['Residence_type'] = df.apply(lambda x: 0 if x['Residence_type'] == 'Urban' else 1, axis=1)
```

```python
df['heart_disease'].unique()
```

```python
df['hypertension'].unique()
```

```python
df['ever_married'].unique()
```

```python
# 0: No, 1: Yes
df['ever_married'] = df.apply(lambda x: 0 if x['ever_married'] == 'No' else 1, axis=1)
```

```python
def check_null_values(df):
    null_df = pd.DataFrame({'columns': df.columns,
                           'percent_null': df.isnull().sum() * 100 / len(df),
                           'percent_zeros': df.isin([0]).sum() * 100 / len(df)})
    return null_df
```

```python
null_df = check_null_values(df)
null_df
```
